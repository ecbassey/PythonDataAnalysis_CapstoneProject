### REVENUE PER MILE ANALYSIS

#### Measures demand

Formula: RPM = number of paying passengers × miles flown

Example:  100 passengers flying 500 miles → 50,000 RPM

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
import nbformat
import plotly.graph_objects as go
```

```python
revenue = pd.read_csv("RPM_rev_per_mile.csv")
```

```python
revenue.head()
```

```python
revenue.info()
```

```python
#convert to datetime format
revenue['observation_date'] = pd.to_datetime(revenue['observation_date'])
```

```python
#Resample (best for time series)
monthly = revenue.set_index('observation_date').resample('M').sum()
```

```python
#revenue['observation_date'] = pd.to_datetime(revenue['observation_date'], format='%y-%b')


revenue['observation_date'] = revenue['observation_date'].dt.to_period('M')
```

```python
#plot chart

fig = px.line(revenue, x='observation_date', y='RPM', title='Airline Revenue Trends')
fig.update_layout(xaxis_title='Month/Year', yaxis_title='Revuenue per mile(Millions)', 
                  width=800,  height=600,)


fig.show()
```

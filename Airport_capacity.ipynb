```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
import nbformat
import plotly.graph_objects as go
```

```python
capacity = pd.read_csv("Airline_Delay_Cause.csv") 
```

```python
capacity.head()
```

```python
capacity.info()
```

```python
filtered = capacity[capacity['year'].isin([2018, 2019, 2020, 2021, 2022, 2023, 2024])]
```

```python
filtered_pre = capacity[capacity['year'].isin([2019])]
```

```python
filtered_post = capacity[capacity['year'].isin([2024])]
```

```python
filtered['airport'].nunique()
```

```python
filtered_pre['airport'].value_counts().head(30
                                            )
```

```python
filtered['city_state'] = filtered['airport_name'].str.split(':').str[0].str.strip()
```

```python
filtered['city_state'].unique()
```

```python
recovery_by_month_pre = filtered_pre.groupby('airport')['arr_flights'].sum().nlargest(20)

recovery_pre = recovery_by_month_pre.reset_index()
```

```python
recovery_by_post = filtered_post.groupby('airport')['arr_flights'].sum().nlargest(30)

recovery_post = recovery_by_post.reset_index()
```

```python
fig = px.bar(
    recovery_pre,
    x='arr_flights',
    y='airport',
    orientation='h',
    #barmode='group',
    color='arr_flights',
    color_continuous_scale='Blues',
    title='Top 10 Airlines by number of flights (Post Covid)',
    labels={'arr_flights': 'Average Flights', 'carrier_name': 'Airline'}
)
fig.update_layout(
    template='plotly_white',
    yaxis_title='Airline',
    xaxis_title='Average Flights',
    title_x=0.5
)
fig.update_layout(height=600)
fig.show()
```

```python
fig = px.bar(
    recovery_post,
    x='arr_flights',
    y='airport',
    orientation='h',
    #barmode='group',
    color='arr_flights',
    color_continuous_scale='Blues',
    title='Top 10 Airlines by number of flights (Post Covid)',
    labels={'arr_flights': 'Average Flights', 'carrier_name': 'Airline'}
)
fig.update_layout(
    template='plotly_white',
    yaxis_title='Airline',
    xaxis_title='Average Flights',
    title_x=0.5
)
fig.update_layout(height=800)
fig.show()
```

```python
comparison = pd.concat([recovery_by_month_pre, recovery_by_post], axis=1)
comparison.columns = ['Pre-COVID', 'Post-COVID']

comparison = comparison.dropna().sort_values(by='Pre-COVID', ascending=False).head(30)
comparison = comparison.reset_index()
```

```python
fig = px.bar(
    comparison,
    x='airport',
    y=['Pre-COVID', 'Post-COVID'],
    barmode='group',
    title='Pre vs Post COVID Airport Traffic Volume (Top 20)',
    labels={'value': 'Average Flights', 'carrier_name': 'Airline'}
)

fig.show()
```

```python
comparison['Recovery Rate'] = comparison['Post-COVID'] / comparison['Pre-COVID']
```

```python
comparison['Recovery Status'] = comparison['Recovery Rate'].apply(
    lambda x: 'Recovered' if x > .95 else 'Not Fully Recovered'
)
```

```python
fig = px.bar(
    comparison.sort_values('Recovery Rate'),
    x='Recovery Rate',
    y='airport',
    orientation='h',
    color='Recovery Status',
    title='Airport Volume Recovery Rate (Post COVID)',
   
)
fig.update_layout(height=650,  yaxis_title="Airport")
fig.show()
```

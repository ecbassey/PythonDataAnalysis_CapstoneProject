```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
import nbformat
```

```python
passenger = pd.read_csv("airline-passenger-load-factor.csv")
```

```python
passenger.head()
```

```python
passenger.info()
```

```python
passenger.dropna(inplace=True)
```

```python
passenger.Year.unique()
```

```python
passenger['Year'] = pd.to_datetime(passenger['Year'], format='%Y')
```

```python
passenger['Seats filled'] = passenger['Seats filled'] / 100
```

```python
fig = px.line(passenger, x='Year', y='Seats filled', markers=True,
              title='Recovery of Passenger Load Factor After COVID-19')
fig.update_layout(xaxis_title='Year', yaxis_title='Percentage of Seats Filled', 
                 yaxis_tickformat='.0%', width=800,  height=600,)


fig.show()
```

```python
covid['Day'] = pd.to_datetime(covid['Day'])
```

```python
#Resample (best for time series)
monthly = covid.set_index('Day').resample('M').sum()
```

```python
covid['New cases (actual)'] = covid['New cases (actual)'] * 1000000
```

```python
fig = px.line(
    covid,
    x='Day',
    y='New cases (actual)',
    title='Covid Cases (Actual Numbers)'
)

fig.show()
```
